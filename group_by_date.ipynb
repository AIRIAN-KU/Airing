### 1. 일치해야 하는 정보 

### a. 출발지→공항 or 공항→목적지 두개 중에 어떤 걸 선택했는지

```python
import pandas as pd

# case1: 공항으로 가는 경우
to_airport_df = pd.read_csv('sample_data/to_airport_data.csv')

# case2: 집으로 가는 경우
to_home_df = pd.read_csv('sample_data/to_home_data.csv')
```

### b. 출발(도착) 날짜가 같아야 함

```python
# case1
to_airport_df['date'] = pd.to_datetime(to_airport_df['date_time']).dt.date
to_airport_group = {date: group.reset_index(drop=True) for date, group in to_airport_df.groupby('date')}
to_airport_group
```

```python
# case2
to_home_df['date'] = pd.to_datetime(to_home_df['date_time']).dt.date
to_home_group = {date: group.reset_index(drop=True) for date, group in to_home_df.groupby('date')}
print(to_home_group)
```
